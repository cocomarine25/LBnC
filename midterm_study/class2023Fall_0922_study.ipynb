## 1. Introduction to Matplotlib for Data Visualization

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

### 1.1 Basic Plotting with Matplotlib

Let's start with a simple line plot:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating the plot
plt.plot(x, y)
plt.title('Simple Line Plot of sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

### 1.2 Bar Plots and Histograms

Bar plots are useful for comparing categories of data:

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values)
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Histograms help in visualizing the distribution of data:

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 2. Working with Arrays and Matrices using NumPy

NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions.

### 2.1 Creating Arrays

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

matrix = np.array([[1, 2], [3, 4], [5, 6]])
```

## 3. Data Manipulation and Analysis with Pandas

Pandas is an open-source data analysis and manipulation tool built on top of Python.

### 3.1 Creating a DataFrame

A DataFrame is a 2-dimensional labeled data structure:

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print(df)
```

### 3.2 Data Selection and Filtering

Selecting a single column:

```python
print(df['Name'])
```

Filtering rows based on conditions:

```python
print(df[df['Age'] > 30])
```

### 3.3 Data Aggregation and Grouping

Grouping data by a column and aggregating:

```python
sales_data = {
    'Region': ['North', 'South', 'North', 'South', 'North', 'South'],
    'Sales': [1000, 1200, 1500, 1800, 1100, 1300]
}

sales_df = pd.DataFrame(sales_data)
print(sales_df.groupby('Region').sum())
```

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Circle
theta = np.linspace(0, 2 * np.pi, 100)
x1 = np.cos(theta)
y1 = np.sin(theta)

# 3D Circle
x2 = np.cos(theta)
y2 = np.sin(theta)
z2 = np.zeros_like(theta)

# 3D Sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x3 = np.outer(np.cos(u), np.sin(v))
y3 = np.outer(np.sin(u), np.sin(v))
z3 = np.outer(np.ones(np.size(u)), np.cos(v))

fig = plt.figure(figsize=(10, 5))

# Circle plot
ax1 = fig.add_subplot(131)
ax1.plot(x1, y1)
ax1.set_title('Circle')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.axis('equal')

# 3D Circle plot
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot(x2, y2, z2)
ax2.set_title('3D Circle')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')

# 3D Sphere plot
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(x3, y3, z3, color='b')
ax3.set_title('3D Sphere')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('z')

plt.tight_layout()
plt.show()
````

````python
# 3D Circle
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

circle = np.linspace(0, 2 * np.pi, 100)
x = np.cos(circle)
y = np.sin(circle)
z = np.zeros_like(x)  # 모든 z 좌표를 0으로 설정합니다.

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # (행, 열, 인덱스)가 (1, 1, 1)이라는 뜻이다.

ax.plot(x, y, z)
ax.set_title('Circle in 3D')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
````

````python
# 3D Sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x3 = np.outer(np.cos(u), np.sin(v))
y3 = np.outer(np.sin(u), np.sin(v))
z3 = np.outer(np.ones(np.size(u)), np.cos(v))

fig = plt.figure(figsize=(10, 5))

# 3D Sphere plot
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(x3, y3, z3, color='b')
ax3.set_title('3D Sphere')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('z')

plt.tight_layout()
plt.show()
````
