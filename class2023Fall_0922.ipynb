## 1. Introduction to Matplotlib for Data Visualization

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

### 1.1 Basic Plotting with Matplotlib

Let's start with a simple line plot:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating the plot
plt.plot(x, y)
plt.title('Simple Line Plot of sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

### 1.2 Bar Plots and Histograms

Bar plots are useful for comparing categories of data:

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values)
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

Histograms help in visualizing the distribution of data:

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## 2. Working with Arrays and Matrices using NumPy

NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions.

### 2.1 Creating Arrays

```python
import numpy as np

# Creating a simple array
arr = np.array([1, 2, 3, 4, 5])
print(arr)

# Creating a 2D array (matrix)
matrix = np.array([[1, 2], [3, 4], [5, 6]])
print(matrix)
```

### 2.2 Array Operations

Basic operations are elementwise:

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Addition:", a + b)
print("Multiplication:", a * b)
```

### 2.3 Mathematical Functions

NumPy provides a host of mathematical functions:

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
```

## 3. Data Manipulation and Analysis with Pandas

Pandas is an open-source data analysis and manipulation tool built on top of Python.

### 3.1 Creating a DataFrame

A DataFrame is a 2-dimensional labeled data structure:

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print(df)
```

### 3.2 Data Selection and Filtering

Selecting a single column:

```python
print(df['Name'])
```

Filtering rows based on conditions:

```python
print(df[df['Age'] > 30])
```

### 3.3 Data Aggregation and Grouping

Grouping data by a column and aggregating:

```python
sales_data = {
    'Region': ['North', 'South', 'North', 'South', 'North', 'South'],
    'Sales': [1000, 1200, 1500, 1800, 1100, 1300]
}

sales_df = pd.DataFrame(sales_data)
print(sales_df.groupby('Region').sum())
```

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

circle = np.linspace(0, 2 * np.pi, 100)
x = np.cos(circle)
y = np.sin(circle)
z = np.zeros_like(x)  # 모든 z 좌표를 0으로 설정합니다.

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z)
ax.set_title('Circle in 3D')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
````
